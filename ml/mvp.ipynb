```python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer('all-MiniLM-L6-v2')  # small, fast
text = "I've been feeling hopeless, tired, and can't focus lately."
embedding = embedder.encode(text)
```

```python
from transformers import pipeline

emo = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
emo(text)
```
